```python
# KDE Plot
# Create a kde plot of sepal_length versus sepal width for setosa species of flower.
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sub = iris[iris["species"] == "setosa"]

sns.kdeplot(
    x="sepal_length",
    y="sepal_width",
    data=sub,
    fill=True,
    cmap="plasma"
)

plt.title("Iris-setosa")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.show()
```

```python
# Dashboard
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')

iris = sns.load_dataset("iris")

f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Boxplot
sns.boxplot(
    x="species",
    y="petal_length",
    data=iris,
    ax=axes[0, 0]
)

# Violin plot
sns.violinplot(
    x="species",
    y="petal_length",
    data=iris,
    ax=axes[0, 1]
)

# Strip plot
sns.stripplot(
    x="species",
    y="sepal_length",
    data=iris,
    jitter=True,
    edgecolor='gray',
    size=6,
    ax=axes[1, 0]
)

# Histogram
axes[1, 1].hist(iris['petal_length'], bins=30)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Clean column names
iris.columns = iris.columns.str.strip()

# Check correct name first
print(iris.columns)

list1 = []
mylabels = []

iris['species'] = iris['species'].astype('category')

for gen in iris['species'].cat.categories:
    list1.append(iris.loc[iris['species'] == gen, 'sepal_length'].values)
    mylabels.append(gen)

plt.hist(list1, bins=30, stacked=True, label=mylabels)
plt.legend()
plt.show()
```
